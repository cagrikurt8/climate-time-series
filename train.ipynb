```python
import pandas as pd
```

## Feature Creation

```python
def create_features(df):
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.day_of_week
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.dayofyear
    return df
```

## Data Exploration

```python
train_df = pd.read_csv("./data/DailyDelhiClimateTrain.csv", index_col='date')
train_df.index = pd.to_datetime(train_df.index)
train_df = create_features(train_df)

test_df = pd.read_csv("./data/DailyDelhiClimateTest.csv", index_col='date')
test_df.index = pd.to_datetime(test_df.index)
test_df = create_features(test_df)
```

```python
df = pd.concat([train_df, test_df])
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_df.drop("humidity", axis=1), test_df.drop("humidity", axis=1), train_df["humidity"], test_df['humidity']
```

## Model Training

```python
import xgboost as xgb
from sklearn.ensemble import RandomForestRegressor

#model = xgb.XGBRegressor(n_estimators=1000, early_stopping_rounds=50, learning_rate=0.01)
model = RandomForestRegressor(n_estimators=1000, verbose=100)
```

```python
#model.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_test, y_test)], verbose=100)
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, pred)
```

## Forecast on Test

```python
test_df['prediction'] = model.predict(X_test)
```
